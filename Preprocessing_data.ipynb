```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dir_path = "Self Driving Car.v3-fixed-small.yolokeras/export/"

with open(os.path.join(dir_path, "_annotations.txt"), "r") as f:
    class_lines = f.readlines()
    # print(len(class_lines[1].strip().split(" "))) 
    # class_names = [line.strip() if len(line.strip().split(" ")) > 1 else None for line in class_lines]
    
    data = []
    for line in class_lines:
        parts = line.strip().split(" ")
        if len(parts) > 1:
            filename = parts[0]
            labels = []
            
            for i in range(1, len(parts[1:])):
                part = parts[i].split(",")
                if part[-1] in ["1"]:
                    labels.append(part)
                    
            if len(labels) > 0:
                data.append((filename, labels))
```

```python
with open("preprocessed_annotations.txt", "w") as f:
    for filename, labels in data:
        label_str = " ".join([",".join(label) for label in labels])
        f.write(f"{filename} {label_str}\n")
```
